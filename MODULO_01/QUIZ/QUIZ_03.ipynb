```python
# Q 17-01

import math 

def distance(x1, y1, x2, y2):
    dx = x2 - y1
    dy = y2 - y1

    distancia = math.sqrt(dx ** 2 + dy ** 2)
    return distancia

x1 = float(input("Introduzca la coordenada x1: "))
y1 = float(input("Introduzca la coordenada y1: "))
x2 = float(input("Introduzca la coordenada x2: "))
y2 = float(input("Introduzca la coordenada y2: "))

resultado = distance(x1, y1, x2, y2)
print(f"La distancia entre los 2 puntos es: {resultado}")
```

```python
# q- 18
def is_palindrome(palabra):
    if len(palabra) <= 1:
        return True
    if palabra[0] != palabra [-1]:
        return False
    return is_palindrome(palabra[1:-1])

palabra = input("Introduzca una palabra: ")
palabra = palabra.lower()

if is_palindrome(palabra):
    print("La palabra es un palíndromo.")
else:
    print("La palabra no es un palíndromo.")
```
