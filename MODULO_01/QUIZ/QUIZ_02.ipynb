```python
# Q12
lst = [5, 6, 3, 9, 2, 12, 3, 8, 7]

for i in range(len(lst) - 1):
    if lst[i] > lst[i + 1]:
        lst[i], lst[i + 1] = lst[i + 1], lst[i]
print(lst)
```
