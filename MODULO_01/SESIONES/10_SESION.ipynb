```python
def suma(n):
    if n==1:
        return 1
    else:
        return n +suma(n-1)
n = int(input("Enter a number: "))
print(suma(n))
```

```python
def potencia(x, n):
    if n == 0:
        return 1
    else: 
        return x * potencia(x, n - 1)
x = int(input("Enter x: "))
n = int(input("Enter n: "))
print(potencia(x,n))
```
