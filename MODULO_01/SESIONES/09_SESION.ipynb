```python
# Q1 pág 343
items = {"coffee": 7, "pen": 3, "paper cup": 2, "milk": 1, "coke": 4, "book": 5}

while True:
    menu = int(input("Select menu 1)check stock 2)warehousing 3)release 4)exit :"))

    if menu == 1:
        item = input("[check stock] Enter item: ").lower()
        if item in items.keys():
            print(f"{item.title()} stock:", items[item])
        else:
            print("Item not found.")

    elif menu == 2:
        item = input("[Warehousing] Enter item and quantity: ").lower()
        quantity = int(input())

        if item in items.keys():
            items[item] += quantity
        else:
            items[item] = quantity

        print(f"{item.title()} updated stock:", items[item])

    elif menu == 3:
        item = input("[Release] Enter item and quantity: ").lower()
        quantity = int(input())

        if item in items.keys():
            if items[item] >= quantity:
                items[item] -= quantity
                print(f"{item.title()} pdated stock:", items[item])
            else:
                print("Not enough stock.")
        else:
            print("Item not found.")

    elif menu == 4:
        print("Program exited.")
        break

    else:
        print("Enter again!")
```

```python
# Q1 pág 403

student_tuple = [
    ("211101", "David Doe", "01-123-1111"),
    ("211102", "John Smith", "010-123-2222"),
    ("211103", "Jane Carter", "010-123-3333"),
]
student_dict = {}
for student in student_tuple:
    student_dict[student[0]] = [student[1], student[2]]

for student_id, student_info in student_dict.items():
    print(f"{student_id}: {student_info[0]}")

student_id = input("Enter student Id number: ")

if student_id in student_dict:
    print(
        f"{student_id} ID student is {student_dict[student_id][0]} "
        f"and phone number is {student_dict[student_id][1]}"
    )
else: 
    print("Student ID incorrect, try again!")
```
