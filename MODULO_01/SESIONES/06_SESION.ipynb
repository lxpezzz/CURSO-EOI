```python
## Q1 PAG 85
s_list = ["abc", "bcd", "bcdefg", "abba", "cddc", "opq"]
corto = s_list[0]
for s in s_list:
    if len(s) < len(corto):
        corto = s
print(f"El texto más corto es {corto}")
```

```python
## Q2 PAG 86
s_list = ["abc", "bcd", "bcdefg", "abba", "cddc", "opq"]
largo = s_list[0]
for s in s_list:
    if len(s) > len(largo):
        largo = s

print(f"El texto más largo es {largo}")
```

```python
##Q3 PAG 87
s_list = ["abc", "bcd", "bcdefg", "abba", "cddc", "opq"]
s_list.sort(key=len)
corto_len = len(s_list[0])
corto = []
for s in s_list:
    if len(s) == corto_len:
        corto.append(s)
print(f"Los textos más cortos:{corto}")
```
