```python
# Q1 pág 529
# Write a program that prints multiplication of 2 using the while statement
# as following.
i = 0
while i <= 9:
    i += 1
    a = 2 * i
    print(f"2 * {i} = {a}")
```

```python
# Q2 PAG 530
# Let's modify the above program to print all the stages 1 to 9 of the 
# multiplication table. USe only the while statement.

i = 1
while i <= 9:
    b = 1
    while b <= 9:
        a = i * b
        print(f"{b} * {i} = {a}")
        b += 1
    i += 1
```

```python
# Q2 PAG 542
# Crear un minijuego en el que hay que adivinar un número entre el 1 y el 100

import random as rd
print("Adivina un número entre el 1 y 100.")
num_random = rd.randint(1, 100)
intentos = 0
while True: 
    num = int(input("Introduce un número: "))
    intentos += 1
    if num > 100:
        print("Elige un número entre el 1 y el 100!")
    elif num == num_random:
        print(f"Has acertado!. Intentos = {intentos}")
        break
    elif num > num_random:
        print("¡Menor!")
    else: 
        print("¡Mayor!")
        


```
