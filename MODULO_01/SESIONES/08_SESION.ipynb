```python
list_2d = []
contador = 1
# Para crear la lista desde 0, sin tener que escribirla a mano
for i in range(1,5):
    fila = []
    for j in range(1, 5):
        fila.append(contador)
        contador += 1   
    list_2d.append(fila)

# Imprimir la lista: 
for fila in list_2d:
    for j in fila:
        print(j, end=" ")
    print()
```

```python
# Q1 PAG 282
n = int(input("Introduzca un número: "))
tablero = []
for i in range(n):
    fila = []
    for j in range(n):
        if (i + j) % 2 == 0:
            fila.append(1)  
        else:
            fila.append(0)
    tablero.append(fila)

for fila in tablero:
    for j in fila:
        print(j, end=" ")
    print()
```
