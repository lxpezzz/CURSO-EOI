```python
#Q1: PAG 343
# Recibimos una puntuación de un usuario, con el condicional if, comprobamos 
# si ha ganado o ha perdido.
score = int(input("Enter game score: "))
if score > 1000:
    print(f"You got {score} points, you win!")
else:
    print(f"You got {score} points, you lose.") 
```

```python
# Q2 PAG 344 
# Write a program that receives any integer x between -100 and 100 and 
# 1) prints x on the screen, and 2) prints "...is a natural number" if 
# x is an integer greater than zero. Otherwise, let it simply print 
# x as in x = -10 
x = int(input("Introduzca un número: "))
if -100 <= x <= 100:
    print(f"x = {x}")
elif x > 0:
    print(f"{x} es un número natural.")
else:
    print("x = -10")


```

```python
# Q1. Write a program tha recieves age as input and prints "Adult" if age
# is 20 or over, "Youth" if under 20 and equal to or over 10, 
# and "Kid" if under 10

edad = int(input("Introduzca un número: "))
if 0 < edad < 10:
    print("Kid")
if edad >= 10 and edad < 20:
    print("Youth")
if edad >= 20:
    print("Adult")
if edad <= 0:
    print("No existes")
```

```python
### TEORIA, si hay in IF dentro de otro IF, para comprobar condiciones 
# específicas.
x = int(input("Introduzca un número: "))
if -100 <= x <= 100:
    if x == 0:
        print("El número es 0.")
    else:
        print(f"x = {x}")
elif x > 0:
    print(f"{x} es un número natural.")
else:
    print("x = -10")
```

```python
## Q2 
# Forma larga de hacerlo:
a = int(input("Introduzca 1 número: "))
b = int(input('Introduzca 1 número: '))
if int(a) % int(b) == 0:
    print(f"{a}, es múltiplo de {b}.")
else:
    print(f"{a} no es múltiplo de {b}.")
```

```python
# Forma corta de hacer el Ejerci

a, b = (input("Introduzca 2 números: ")).split()
if int(a) % int(b) == 0:
    print(f"{a}, es múltiplo de {b}.")
else:
    print(f"{a} no es múltiplo de {b}.")
```
