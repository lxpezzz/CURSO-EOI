```python
## Fórmula de factorial:
"""
0! = 1
k! = k x (k - 1)
"""
def factorial(k):
    if k == 0 or k == 1:
        return 1
    return k * factorial(k - 1)

def euler(n):
    if n == 0:
        return 1 / factorial(0)
    return euler(n - 1) + 1  / factorial(n)

resultado = euler(20)
print(f"Euler(20) = {round(euler(20),5)}")
```

```python
def factorial(k):
    if k == 0 or k == 1:
        return 1
    return k * factorial(k - 1)


def euler(n):
    if n == 0:
        return 1 / factorial(0)
    return euler(n - 1) + 1 / factorial(n)


resultado = euler(20)

print(f"eular(20) = {resultado:.5f}")
```

```python
#Q1 
scores = [100, 90, 95, 90, 80, 70, 0, 80, 90, 90, 0, 90, 100, 75, 20, 30, 50, 90]
total_estudiantes = len(scores) // 3
estudiantes_validos = []

for i in range(0, len(scores), 3):
    estudiante = scores[i : i + 3]

    if 0 not in estudiante:
        estudiantes_validos.append(estudiante)

print(f"scores = {scores}")
print(f"The number of total students is {total_estudiantes}.")
print(f"The number of students with valid scores is {len(estudiantes_validos)}.")
print(estudiantes_validos)
```
