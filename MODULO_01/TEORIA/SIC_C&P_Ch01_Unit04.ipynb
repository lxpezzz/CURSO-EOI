### 2.2. How the BMI CALCULATOR works

```python
name = input("Enter name: ")
height_cm = int(input("Enter height(cm): "))
height_m = height_cm / 100
weight = int(input("Enter weight (kg) : "))
bmi = weight / (height_m ** 2)
print(name, "'s BMI is", bmi)
```
