```python
a = 100
b = 200
print('before swap : a = ', a, 'b = ', b)
temp = a
a = b
b = temp
print('after swap : a = ', a, 'b = ', b)
```

```python
a = 100
b = 200
print('before swap : a = ', a, 'b = ', b)
a, b = b, a    # very simple swap
print('swap result using tuple: a = ', a, 'b = ', b)
```
