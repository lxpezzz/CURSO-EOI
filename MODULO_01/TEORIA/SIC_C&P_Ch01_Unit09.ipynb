### 3. Mission

```python
current_pos = 0  # Current position of snail
days = 1         # date
move_up = 7      # distance it goes up a day : in meters
move_down = 5    # distance it goes down while sleep : in meters
destination = 30 # Distance goal : in meters
current_pos += move_up # position of the snail on the first day

print('day :', days, "Snail's position ", current_pos, 'meters')
while(current_pos <= destination): # repeat 17 snail hasn't reached the distance goal
    current_pos -= move_down       # goes down again
    days += 1                      # next day
    current_pos += move_up         # distance it went up
    print('day :', days, "snail's position :", current_pos, 'meters')
print()
print('It took {} days to get out of the well.'.format(days))
```

### 1. Basic Structure of while Statement

```python
i = 0 # initial value
while i < 5:  # block is executed when loop's condition is true
    print('Welcome to everyone!!')
    i += 1
```

```python
n = int(input('Enter a number to sum up to : ' ))
s = 0
i = 1
while i <= n:
    s = s + i
    i += 1
print ('Sum of 1 to {} is {}'.format(n, s))
```

```python
selected = None
while selected not in ['scissors', 'rock', 'paper']:
    selected = input('Choose among scissors, rock, paper> ')
print('Chosen value:', selected)
```

```python
n = int(input('Enter a number to sum up to :'))
s = 0
for i in range(1, n+1) :
    s = s + i
print ('Sum of 1 to {} is {}'.format(n, s))
```

```python
n = -1
while n <= 0: # input() repeated until positive value is entered
    n = int(input('Enter a positive number to sum up to :'))
s = 0
for i in range(1, n+1):
    s = s + i
print('Sum of 1 to {} is {}'. format(n, s))
```

### 2. random

```python
import random as rd    # import command that calls inner module and abbreviation of random module, rd

print(rd.random ())            # Returns a random real
print(rd.random())             # Returns a random real
print(rd.randrange (1, 7))     # Returns an integer greater than or equal to 1 and less than 7.
print(rd.randrange (0, 10, 2)) # Returns a multiple of 2 of integers greater than or equal to 0 and less than 10.
print(rd.randint(1, 10))       # Returns an arbitrary integer greater than or equal to 1 and less than
                               # or equal to 10 (including10).
```

### 3. break, continue

```python
st = 'Programming'
# function of executing only when it is consonant
for ch in st:
    if ch in ['a', 'e', 'i','o','u']:
        break    # end the loop if it is vowel
    print (ch)
print('The end')
```

```python
st = 'Programming'
# function of executing only when it is consonant
for ch in st:
    if ch in ['a','e','¡','o','u']:
        continue    # skip the execution below if it is vowel
    print(ch)
print ('The end')
```

```python
for i in range (2, 10):            # outer for loop
    for j in range(1, 10) :        #inner for loop
        print('{}x{} = {:2d}, '. format (i, j, i*j), end = ' ')
    print()        # executes inner loop and change line
```

### 1. while Statement

```python
under_construction = True
while under_construction:
    response = input("Is the construction completed?");
    if response == "Yes":
        under_construction = False

print("Escaped successfully.")
```

```python
password = ""
while password != "pythonisfun":
    password = input("Enter the password: ")
print("** Login success **")
```

```python
count = 1
s = 0 # s is variable that contains values accumulated and added, reset it as 0
while count <= 10 :
    s = s + count # add count value to s each time
    count = count + 1 # increase count value by 1 every time
print("Sum is", s)
```

```python
total = 0
answer = 'yes'
while answer == 'yes':
    number = int(input ('Enter the number: '))
    total = total + number
    answer = input ('Continue?(yes/no): ')
print('Sum is : ', total)
```
