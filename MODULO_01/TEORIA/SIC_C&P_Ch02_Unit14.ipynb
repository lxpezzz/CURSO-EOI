### 3. Mission

```python
print("Start café menu program... Press q to exit")
cafe_menu = {'Ice coffee' : 3000}

while True:
    str = input('$ ')
    if str.startswith('q'):
        break;
    command = str[0]

    if command == '<':
        str = str.replace('<', "")
        inputStr = str.split(':')
        if len(inputStr) < 2:
            print('input error')
            continue
        else:
            cafe_menu[inputStr[0].strip()] = inputStr[1].strip()
    
    elif command == '>':
        str = str. replace('>',"")
        inputStr = str.strip()
        if inputStr in cafe_menu:
            print(cafe_menu[inputStr])
        else:
            print('{} not in the menu.'.format(inputStr))
    elif command == 'q':
        break
    else:
        print('input error.')
print("exiting cafe menu program.")
```

## Key concept

### 1. Add and Edit Key-Value Together

```python
str1 = ['a', 'b', 'c', 'b', 'a', 'b', 'c']
dic = {}
for ch in str1:
    if ch not in dic.keys():
        dic[ch] = 0
    dic[ch] += 1

print('alphabet counting :', dic)
```

```python
str1 = ['a', 'b', 'c', 'b', 'a', 'b', 'c']
dic = {}
for ch in str1:
    dic.setdefault(ch, 0)
    dic[ch] += 1
    print(list(dic.items()))

print('alphabet counting :', dic)
```
