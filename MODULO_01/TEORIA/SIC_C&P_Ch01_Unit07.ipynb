### 2. Nested Conditional Statement

```python
num = int(input("Enter an integer: "))
if num > 0:
    print("It is positive.")
elif num == 0:
    print("It is 0.")
else:
    print("It is negative.")
```

```python
score = int (input ('Enter score : '))
if score >= 90:                # If 90 or above, 'A'
    grade = 'д'
if score < 90 and score >= 80: # If under 90, 80 or above, 'B'
    grade = 'B'
if score < 80 and score >= 70: # If under 80, 70 or above,'C'
    grade = 'C'
if score < 70 and score >= 60: # If under 70, 60 or above, 'D'
    grade = 'D'
if score < 60:                 #If under 60, 'F'
    grade = 'F'
print ('Your grade is: ', grade)
```

```python
score = int(input ('Enter score : '))
if score >= 90:                # If 90 or above, 'A'
    grade = 'A'
else:
    if score >= 80:            # If under 90, 80 or above, 'B'
        grade = 'B'
    else:
        if score >= 70:        # If under 80, 70 or above, 'C'
            grade = 'c'
        else:
            if score >= 60:   # If under 70, 60 or above, 'D'
                grade = 'D'
            else:
                grade = 'F'   #If under 60, 'F'

print('Your grade is: ', grade)
```

```python
score = int(input('Enter score : '))
if score >= 90:      # If 90 or above, 'A'
    grade = 'A'
elif score >= 80:    # If not 'A', 'B' if 80 or above
    grade = 'B'
elif score >= 70:    # If not 'B', 'C' if 70 or above
    grade = 'C'
elif score >= 60:    # If not 'C', 'D' if 60 or above
    grade = 'D'
else:                # Otherwise, 'F'
    grade = 'F'
print('Your grade is: ', grade)
```

```python
a = 10
b = 14                # if b is changed to 13, it does not meet the first conditional statement
if (a % 2 == 0) and (b % 2 == 0):    # first conditional statement
    print('Both are even.')
if (a % 2 == 0) or (b % 2 == 0):     # second conditional statement
    print('At least one is even.')
```

### 3. Application production practice

```python
print('#' * 50)
print("This is David's fruit shop.")
print('1: Apple( price: 5000 won)')
print('2: Grape( price: 6000 won)')
print('3: Melon( price : 8000 won)')
print('4: Orange( price : 2000 won)')
print('#' * 50)

total_price = 0
order = int(input(('Enter item number (between 1~4) >> ')))
count = int(input(('Enter number of items (between 1~10) >> ')))

if order == 1:
    fruit = 'Apple'
    price = 5000
elif order == 2:
    fruit = 'Grape'
    price = 6000
elif order == 3:
    fruit = 'Melon'
    price = 8000
else:
    fruit = 'Orange'
    price = 2000

print('Fruit selected :', fruit)
print('Price.:', price)
print('Quantity :', count)
print('Total price is', price * count, 'won.')

money = int(input('Insert money please(ex: 15000) >>>'))
if money < price * count :
    print('Not enough money.')
else:
    change = money - price * count
    print(money, 'won received. Your change is', change, 'won.')
```
