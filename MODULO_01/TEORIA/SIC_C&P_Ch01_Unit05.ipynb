## Let’s code

### 1. Logical Operators: AND, OR, NOT

```python
print(10 > 20) # 10 is less than 20, so 10 > 20 is False
print(10 < 20) # 10 is less than 20, so 10 < 20 is True
print(bool (9)) # 9 is not 0, so it is True
print(bool (-1)) # -1 too is not 0, so it is True
print(bool(0)) # 0 is the only number value that is False
print(bool (None)) # None means there is no value, so it is False
print(bool('')) # White space is a string, so it is False
print(bool('hello')) # It holds string values, so it is True
```
