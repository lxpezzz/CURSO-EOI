```python
# Ejercicio 11
list_numero = []
contador = 0
for i in range(5):
    numero = int(input("Introduzca un número: "))
    list_numero.append(numero)
    if numero < 0:
        contador += 1
print(f"El hay {contador} números negativos en tu lista")
print(list_numero)
```

```python
# Ejercicio 12
nombres = []

for i in range(5):
    nombre = str(input("Introduzca un nombre: "))
    nombres.append(nombre)

mas_largo = nombres[0]

for nombre in nombres: 
    if len(nombre) > len(mas_largo):
        mas_largo = nombre
print(mas_largo)
```

```python
# Ejercicio 13
cadena = (input("Introduzca un número de más de 3 dígitos: "))
lista = cadena.split()
suma = 0
for n in lista:
    suma = suma + int(n)
print(suma)
```

```python
# Ejercicio 14
list_numeros = []
nueva_lista = []
for i in range(8):
    numero = int(input("Introduzca un número: "))
    list_numeros.append(numero)
for numero in list_numeros:
    if numero not in nueva_lista:
        nueva_lista.append(numero)
print(nueva_lista)

```

```python
# Ejercicio 16
lista_numeros = []
for i in range(6):
    numeros = int(input("Introduzca un número: "))
    lista_numeros.append(numeros)

lista_numeros.sort()

mayor = max(lista_numeros)
lista_numeros.remove(mayor)

segundo_mayor = max(lista_numeros)
print(segundo_mayor)
```

```python
# Ejercicio 17
lista_01 = []
lista_02 = []
for i in range(4):
    numeros = int(input("Introduzca un número: "))
    lista_01.append(numeros)
for j in range(4):
    numeros = int(input("Introduzca un número: "))
    lista_02.append(numeros)
suma_lista = []
for i in range(len(lista_01)):
    suma = lista_01[i] + lista_02[j]
    suma_lista.append(suma)
print(suma_lista)
```

```python
# Ejercicio 18
num_10 = []

for i in range(10):
    numero = int(input("Introduzca diez números: "))
    num_10.append(numero)
suma = 0
for numero in num_10:
    suma += numero
media = suma / len(num_10)
contador = 0
for numero in num_10:
    if numero > media:
        contador += 1
print(contador)
```
