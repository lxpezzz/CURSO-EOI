```python
# TAREA PARA ENTREGAR:
# Q1 PÁG 420
'''Write a program that executes addition, substraction, multiplication, and division.
It prints the operation result of two positve integers, based on the desired
operation number given as input. If a number else than 1, 2, 3, or 4 is given as input, 
the program should print an error message. To enter two numbers, write one, press
Enter, and then write the other.'''

print("Esto es un programa para sumar, restar, multiplicar y dividir, "
      "elija 1 de las 4 opciones, de 1, 2, 3, 4; en ese debido orden\n")

op= int(input("Introduzca el número de operaciones: "))
num_1 = int(input("Introduzca el 1º número: "))
num_2 = int(input("Introduzca el 2º número: "))

if op == 1:
    print(num_1 + num_2)
elif op == 2:
    print(num_1 - num_2)
elif op == 3:
    print(num_1 * num_2)
elif op == 4:
    print(num_1 // num_2)
else:
    print("Has introducido una operación incorrecta, vuelve a probar!")
```

```python
# Q2. Pág 421
# Write a program that receives a point with x and y coordinates as input,
# and determines in which quadrant among 1, 2, 3, 4 the point belongs. The
# position of the quadrant is shown in the following figure.
"""
1º Cuadrante 'x' positivo y 'y' positivo.
2º Cuadrante 'x' negativo y 'y' positivo
3º Cuadrante 'x' negativo y 'y' negativo
4º Cuadrante 'x' positivo y 'y' negativo
"""

coord_x = int(input("Introduzca la coordenada 'x': "))
coord_y = int(input("Introduzca la coordenada 'y': "))
if coord_x > 0 and coord_y > 0: 
    print("En el 1º cuadrante.")
elif coord_x < 0 and coord_y > 0:
    print("En el 2º cuadrante.")
elif coord_x < 0 and coord_y < 0:
    print("En el 3º cuadrante.")
elif coord_x > 0 and coord_y < 0:
    print("En el 4º cuadrante.")
else:
    print("Tienes una coordenada con valor 0!")
```

```python
# Q3. PAG 422
# Develop a menu ordering program for Yummy Restaurant .
# Show the following menu to the user and let the user select one.
# If the given input alphabet is not in the menu, 'Enter the menu again:' and
# Receive another input.
lista_menu = ['hamburguesa', 'pollo', 'pizza']
print("Bienvenido a Yummy Restaurant. Aquí tienes el menú\n")
for plato in lista_menu:
    print(f"- {plato.title()}")

menu = input("Introduzca el menú: ")

while menu not in lista_menu:
    print("Introduzca el menú otra vez: ")
    menu = str(input("Introduzca el menú (escribe 'quit' para salir): "))
    if menu == 'quit':
        break       
if menu == 'hamburguesa':
    print(f"Has elegido el menú: {menu.title()}.")
elif menu == 'pollo':
    print(f"Has elegido el menú: {menu.title()}.")
elif menu == "pizza":
    print(f"Has elegido el menú: {menu.title()}.")
```
