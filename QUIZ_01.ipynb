```python

for num in range(2, 13):
    primo = True
    for divisor in range(2, num):
        if num % divisor == 0:
            primo = False
            break
    if primo:
        print(f"El número {num} es un número primo.")
    else:
        print(f"El número {num} es un número compuesto.")
```

```python
for num_arms in range(100, 1000):
    centenas = num_arms // 100
    decenas = (num_arms % 100) // 10
    unidades = num_arms % 10

    if num_arms == centenas**3 + decenas**3 + unidades**3:
        print(f"{num_arms} es un número Armstrong")

## 153 = 1 + 125 + 27
## 370 = 27 + 343 + 0
## 371 = 27 + 343 + 1
## 407 = 64 + 0 + 343
```
