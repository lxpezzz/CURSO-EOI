```python
def josephus_sequence(n, k):
    sequence = []
    queue = []
    for i in range(1, n + 1):
        queue.append(i)
    j = 1
    while len(queue) > 1:
        item = queue.pop(0)
        if i % k == 0:
            sequence.append(item)
        else:
            queue.append(item)
        i += 1
    sequence.append(queue.pop(0))
    return sequence
```

### 2. Implement a Queue

```python
class Queue:
    
    def __init__(self):
        self.queue = []
    
    def is_empty (self):
        return True if len(self.queue) == 0 else False
    
    def enqueue (self, item):
        self.queue.append(item)
    
    def dequeue(self):
        return None if self.is_empty() else self.queue.pop(0)
```

### 1. Josephus Problem

```python
def josephus(n, k):
    sequence = []
    queue = Queue()
    for i in range(1, n + 1):
        queue.enqueue(i)
    j = 1
    while queue.size() > 1:
        item = queue.dequeue()
        if j % k == 0:
            sequence.append(item)
        else:
            queue.enqueue(item)
        j += 1
    sequence.append(queue.dequeue())
    return sequence
```

```python
N = int(input("Input the number of people(N): "))
K = int(input ("Input the number to be skipped (K) : "))
print (josephus(N, K))
```
