```python
# Q2 PAG 61 ALGORTIMOS
from estructuras import Stack

stack = Stack()
items = [10* i for i in range(1, 10)]
for item in items:
    stack.push(item)
    if (item//10) % 2 == 0:
        stack.pop()
print(stack.stack)
```

```python
from estructuras import Queue 
queue = Queue()
items = [10 * i for i in range(1, 11)]
for item in items:
    queue.enqueue(item)
    if (item // 10) % 2 == 0:
        queue.dequeue()
print(queue.queue)
```

```python
class Deque:  
    def __init__(self):
        self.queue = []
    def is_empty(self): # creamos esta función para decir si está vacía la cola.
        return len(self.queue) == 0    
    def add_first(self, item):
        '''Add an item to the front of the queue'''
        self.queue.insert(0, item)
    def remove_first(self):
        '''Remove and return the item from the front'''
        if self.is_empty() == 0:
            return None   #  si la cola está vacía devuelvo None al eliminar
        return self.queue.pop(0)
    
    def add_first(self, item):
        '''Add an item to the rear of the queue'''
        self.queue.append(item)
    def remove_first(self):
        '''Remove and return the item from the rear'''
        if self.is_empty() == 0: 
            return None
        return self.queue.pop()
```

```python
# quiz 1 pag 174
def find_two(nums):
    x = y = 0 
    for i in range(1, len(nums)):
        if nums[x] < nums[i]:
            x = i
        elif nums[y] > nums [i]:
            y = i
    return x, y
nums = [11, 37, 45, 26, 59, 28, 17, 53]
i, j = find_two(nums)
print(nums[i], nums[j])
```

```python
# q1 pag 179
def word_count(S, x):
    count = 0
    for word in S:
        if word == x:
            count += 1
    return count
S = input("Input a sentence: ").split()
x = input("Input a word to search: ")

count = word_count(S, x)

print(f"In S, {x} is appeared in {count} times.")
```
