```python
# Operación entre 2 elementos
#  S = Lista
# x = 1º elemento
# y = 2º elemento

def swap1(S, x, y):
    t = S[x]
    S[x] = S[y]
    S[y] = t
S = [10, 20]
print(S)
swap1(S, 0, 1)
print(S)
```

```python
# Versión simplificada
def swap1(S, x, y):
    S[x], S[y] = S[y], S[x]
S = [10, 20]
print(S)
swap1(S, 0, 1)
print(S)
```

```python
# Bubble Sort
def bubblesort(S):
    n = len(S)
    for i in range(n):
        print(S)
        for j in range(n - 1):
            if S[j] > S[j + 1]:
                S[j], S[j + 1] = S[j + 1], S[j]

S = [20, 32, 70, 14, 35]
bubblesort(S)
```

```python
# Q1 71
def son_anagramas(palabra1, palabra2):
    palabra1 = palabra1.lower()
    palabra2 = palabra2.lower()
    if len(palabra1) != len(palabra2):
        return False
    if sorted(palabra1) == sorted(palabra2):
        return True
    else:
        return False
print(son_anagramas("LISTEN", "SILENT"))
```
